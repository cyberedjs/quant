### 결측치와 이상치 다루기

```python
import numpy as np
s1 = pd.Series([1, np.nan, 3, 4, 5])
s2 = pd.Series([1, 2, np.nan, 4, 5])
s3 = pd.Series([1, 2, 3, np.nan, 5])
df = pd.DataFrame({'S1':s1,
                   'S2':s2,
                   'S3':s3
                  })
df.head()
```

```python
df['S1'].isna()
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isin([np.nan])
```

```python
df.fillna(0)
```

```python
df.fillna(method = 'pad')
```

```python
df.fillna(method = "bfill")
```

```python
df.dropna()
df.dropna(axis='rows')
df.dropna(axis=0)
```
